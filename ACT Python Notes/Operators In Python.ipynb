# Comparison Operator

```python
x = int(input("Enter the number: "))
y = int(input("Enter the number: "))

if(x>y):
    print("x is big")
elif(x<y):
    print("y is big")
elif(x==y):
    print("Both are same")
else:
    print("Null")
```

```python
q = int(input("Enter the number: "))
w = int(input("Enter the number: "))

if(q != w):
    print("Not same")
elif(q <= w):
    print("x is small")
elif(q >= w):
    print("y is small")
else:
    print("Null")
```

# Logical Operator

```python
#And
x = 1
y = 0
print("x and y is: ",x and y)

#Or
x = 0
y = 1
print("x or y is: ",x or y)

#Not
x = 0
print("not x is: ",(not x))
```

# Bitwise Operator

```python
# & | ~ ^ >> <<

i = 10
j = 4

print("i & j: ",i & j)
print("i | j: ", i | j)
print(" ~ i: ", ~i)
print("i^j: ", i^j)
print("i>>2: ", i>>2)
print("i<<2", i<<2)
```
