```python
import networkx as nx
```

```python
import matplotlib.pyplot as plt
```

```python
G=nx.Graph()
```

```python
G.add_node(1)
G.add_node(2)
G.add_node(3)
G.add_node(4)
G.add_node(5)
```

```python
G.nodes()
```

```python
G.add_edge(1,2)
G.add_edge(2,3)
G.add_edge(3,4)
G.add_edge(1,4)
G.add_edge(1,5)
```

```python
G.edges()
```

```python
nx.draw(G,with_labels=1)
```

```python
z=nx.complete_graph(10)
```

```python
z.nodes()
```

```python
z.edges() #45 different combinations possible for 10 nodes
```

```python
z.order() #number of nodes in the graph 
z.size() #number of edges in the graph
```

```python
nx.draw(z)
plt.savefig("abc1.png")
#plt.show()
```
