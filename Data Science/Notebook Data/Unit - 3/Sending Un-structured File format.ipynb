```python
from skimage.io import imread
from skimage.transform import resize
from matplotlib import pyplot as plt
import matplotlib.cm as cm
```

```python
example_file = ("RGB_image.jpg")
image = imread(example_file)
plt.imshow(image)
plt.show()
```

```python
# Know the type of image
print(type(image))
```

```python
#image2 = image[0:225,0:225,2] 
image2 = image[5:150,0:170,2]           #first is vertical and second is horizontal axis
plt.imshow(image2)
plt.show()
```

```python
plt.imsave("cropped_RGB.jpg",image2)
```

```python
image3 = resize(image, (200, 20), mode='symmetric')
plt.imshow(image3)
print("data type: %s, shape: %s" % (type(image3), image3.shape))
```

```python
image_row = image3.flatten()
print("data type: %s, shape: %s" %
(type(image_row), image_row.shape))
```

The image is currently an
array of 30 pixels by 30 pixels, so you cant make it part of a dataset. The following
code fattens image3 so that it becomes an array of 900 elements that is stored in
image_row.Notice that the type is still a numpy.ndarray. You can add this array to a dataset
and then use the dataset for analysis purposes. The size is 900 elements, as
anticipated.

```python
from PIL import Image
 
# Opens a image in RGB mode
im = Image.open("RGB_image.jpg")
 
# Size of the image in pixels (size of original image)
# (This is not mandatory)
width, height = im.size
 
# Setting the points for cropped image
left = 5
top = height / 4
right = 164
bottom = 3 * height / 4
 
# Cropped image of above dimension
# (It will not change original image)
im1 = im.crop((left, top, right, bottom))
 
# Shows the image in image viewer

plt.imshow(im1)
plt.show()
```
